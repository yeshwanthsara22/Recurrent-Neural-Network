```python
import numpy as np
import keras
import random
import glob
import os
import tensorflow_addons as tfa

class GRUTest:
    def __init__(self, model_path="GRU_Model.h5", test_data_path="/Users/yeshwanth/Documents/DL/project-2/part-1/test.csv"):
        self.model_path = model_path
        self.test_data_path = test_data_path
    
    def load_model(self):
        custom_objects = {
    'PeepholeLSTMCell': tfa.rnn.PeepholeLSTMCell
}
        model=None
        with keras.utils.custom_object_scope(custom_objects):
        # Load the trained model
            model = keras.models.load_model(self.model_path)
        # model = keras.models.load_model(self.model_path)
        return model
    def load_file(self,file_path,train_x_list):
        l=[]
        with open(file_path, 'r') as f:
            data = []
            c=0
            for line in f:
                c=c+1
                # print(line)
                if '#' in line:
                    continue  # Skip comment lines
                    # print(line)
                line_data = line.strip().split(',')  # Skip the first column and split remaining data
                # print(line_data)
                l.append(line_data)
                numeric_data = [float(x) for x in line_data if x.strip() and x.strip() != '#']  # Convert non-empty strings to float
                if len(numeric_data)==4:
                    numeric_data=numeric_data[:3]
                if len(numeric_data)==5:
                    numeric_data=numeric_data[1:4]
                
                data.append(numeric_data)
                if c==31 or c%31==0:
                    data = np.array(data)
                    train_x_list.append(data)
                    data=[]


# Function to load data from all files in a directory
    def load_data_from_directory(self):
        file_paths = glob.glob(self.test_data_path)
        csv_paths = [file_path for file_path in file_paths if os.path.splitext(file_path)[1].lower() == ".csv"]
        if len(csv_paths)==0:
            file_paths = glob.glob(self.test_data_path+"/**/*.csv")
        # print(file_paths)
        random.shuffle(file_paths)
        train_x_list = []
        for file_path in file_paths:
            self.load_file(file_path,train_x_list)
        return np.array(train_x_list)

    def predict(self, model,x_test):
        # print(model.summary())
        predictions = model.predict(x_test)
        max_indices = np.argmax(predictions, axis=1)
        return max_indices

# Example usage:
if __name__ == "__main__":
    gru_test = GRUTest(model_path="/Users/yeshwanth/Documents/DL/project-2/part-2/LSTM_Model.h5", test_data_path="test.csv")
    model = gru_test.load_model()
    x_test = gru_test.load_data_from_directory()
    # print(x_test)
    predictions = gru_test.predict(model, x_test)
    print(predictions)
   
```
